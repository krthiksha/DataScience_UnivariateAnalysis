# Data Preprocessing - Chronic Kidney Disease

```python
import pandas as pd 
dataset = pd.read_csv("kidney_disease.csv")
dataset
```

```python
dataset.info()
```

## finding missing values

```python
# find missing values - mv
mv_details = dataset.isnull().sum()
mv_details
```

```python
# quan qual 
from Univariate_analysis import Univariate
Univariate = Univariate(dataset)
quan , qual = Univariate.quanQual()
quan=quan[1:]
print("quan :",quan , "\nqual :", qual)
```

## replacing missing values

```python
dataset[quan]
```

```python
####  replace quan missing values with simple Imputer 

import numpy as np 
import pandas as pd
from sklearn.impute import SimpleImputer 
imp = SimpleImputer(missing_values = np.nan, strategy='median')
imp.fit(dataset[quan])
quan_dataset = imp.transform(dataset[quan])
quan_dataset=pd.DataFrame(quan_dataset, columns=quan)
quan_dataset
```

```python
# replace qual missing values with most repeated data
# approach - 1 
# for column in mv_col_qual:
#     dataset.fillna({column:dataset[column].mode()[0]},inplace=True)
# dataset

# replace qual missing values with 'unknown' - constant string
# approach - 2
import numpy as np 
import pandas as pd
from sklearn.impute import SimpleImputer 
imp = SimpleImputer(missing_values = np.nan, strategy='constant', fill_value='unknown')
imp.fit(dataset[qual])
qual_dataset = imp.transform(dataset[qual])
qual_dataset=pd.DataFrame(qual_dataset,columns=qual)
qual_dataset
```

##  concat both dataframe

```python
preprocessed_dataset = pd.concat([quan_dataset,qual_dataset], axis=1)    # concat axis - 1 (column)
preprocessed_dataset
```

```python
preprocessed_dataset.isnull().sum()
```

```python
# preprocessed_dataset['classification'].replace({'ckd':1,'notckd':0},inplace=True)
```

```python
preprocessed_dataset['classification'] = preprocessed_dataset['classification'].str.strip()
preprocessed_dataset['pcv'] = preprocessed_dataset['pcv'].str.strip()
preprocessed_dataset['wc'] = preprocessed_dataset['wc'].str.strip()
preprocessed_dataset['dm'] = preprocessed_dataset['dm'].str.strip()
preprocessed_dataset['cad'] = preprocessed_dataset['cad'].str.strip()
```

```python
preprocessed_dataset['wc'] = preprocessed_dataset['wc'].replace(['?','unknown'],'0')
preprocessed_dataset['wc'].unique()
```

```python
preprocessed_dataset['pcv'] = preprocessed_dataset['pcv'].replace(['?','unknown'],'0')
```

```python
preprocessed_dataset['pcv'].unique()
```

```python
preprocessed_dataset['rc'] = preprocessed_dataset['rc'].replace(['\t?','unknown'],'0')
```

```python
preprocessed_dataset['rc'].unique()
```

```python
preprocessed_dataset['rc']=preprocessed_dataset['rc'].astype('float64')
preprocessed_dataset['pcv']=preprocessed_dataset['pcv'].astype('int64')
preprocessed_dataset['wc']=preprocessed_dataset['wc'].astype('int64')
```

```python
preprocessed_dataset.info()
```
